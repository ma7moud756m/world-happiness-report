```python
dfs ={
    2015 : pd.read_csv(r"D:\ME\data analsis\projects\project 7 (happeniss) 2\Data\2015.csv"),
    2016 : pd.read_csv(r"D:\ME\data analsis\projects\project 7 (happeniss) 2\Data\2016.csv"),
    2017 : pd.read_csv(r"D:\ME\data analsis\projects\project 7 (happeniss) 2\Data\2017.csv"),
    2018 : pd.read_csv(r"D:\ME\data analsis\projects\project 7 (happeniss) 2\Data\2018.csv"),
    2019 : pd.read_csv(r"D:\ME\data analsis\projects\project 7 (happeniss) 2\Data\2019.csv")
}
```

```python
rename = {
     "Country or region": "Country",
     "Country": "Country",

     'Score': 'Happiness Score',
     'Happiness.Score': 'Happiness Score',

     'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
     'GDP per capita': 'Economy (GDP per Capita)',

     "Social support": 'Family',
     "Family": "Family",

     'Health..Life.Expectancy.': 'Health (Life Expectancy)',
     'Healthy life expectancy': 'Health (Life Expectancy)',

     'Freedom to make life choices': 'Freedom',

     'Trust..Government.Corruption.': 'Trust (Government Corruption)',
     'Perceptions of corruption': 'Trust (Government Corruption)',

     'Generosity': 'Generosity'
}
```

```python
cols = ["Country", "Happiness Score", "Economy (GDP per Capita)",
        "Family", "Health (Life Expectancy)", "Freedom",
        "Trust (Government Corruption)", "Generosity"]
```

```python

combined = []
for year, df in dfs.items():
    df = df.rename(columns=rename)
    df = df[cols]
    df["Year"] = year
    combined.append(df)
```

```python
df_all = pd.concat(combined, ignore_index=True)
```

```python
df_all
```

```python
df = df_all.copy()
```

```python
df.info()
```

```python
plt.bar(
    df['Year'] , df['Happiness Score'] , 
    
)
```

```python
def is_happy (x):
    if x >= 5:
        return True
    else:
        return False

df['is_happy'] = df['Happiness Score'].apply(is_happy)
```

```python
labels = ['Happy', "Unhappy"]
plt.pie(
    df['is_happy'].value_counts(),
    labels=labels,
    autopct="%1.0f%%"
)
plt.axis = ['equal']
```

```python
top10 = (df.groupby("Country")["Happiness Score"]
    .mean()
    .nlargest(10)
)
top10
```

```python
top10.sort_values().plot(kind = 'barh')
```

```python
df_top10 = df[df["Country"].isin(top10.index)]
```

```python
df.info()
```

```python
pivot_df = df_top10.pivot_table(index="Year", columns="Country", values="Happiness Score",aggfunc="mean")
# pivot_df.head()
pivot_df.plot()

plt.title("Top 10 Countries")
plt.ylabel("Happiness Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.figure(figsize=(10,6))
plt.tight_layout()
plt.show()
```

```python
top10 = (df.groupby("Country")["Economy (GDP per Capita)"]
    .mean()
    .nlargest(10)
)
top10.sort_index()
df_top10 = df[df["Country"].isin(top10.index)]

```

```python
pivot_df = df_top10.pivot_table(index="Year", columns="Country", values="Economy (GDP per Capita)",aggfunc="mean")
# pivot_df.head()
pivot_df.plot()

plt.title("Top 10 Countries")
plt.ylabel("Economy (GDP per Capita)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.figure(figsize=(10,6))
plt.tight_layout()
```

```python
lowest10 = (df.groupby("Country")["Happiness Score"]
    .mean()
    .nsmallest(10)
)
```

```python
df_lowest10 = df[df["Country"].isin(lowest10.index)]
```

```python
pivot_df = df_lowest10.pivot_table(index="Year", columns="Country", values="Happiness Score",aggfunc="mean")
# pivot_df.head()
pivot_df.plot()

plt.title("lowest 10 Countries")
plt.ylabel("Happiness Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.figure(figsize=(10,6))
plt.tight_layout()
```

```python
df_15 = df[df['Year']==2015][['Country' , 'Happiness Score']]
df_19 = df[df['Year']==2019][['Country' , 'Happiness Score']]
Score = df_15.merge(df_19, on='Country', suffixes=('_2015', '_2019'))
Score['Change'] = Score['Happiness Score_2019'] - Score['Happiness Score_2015']
```

```python
topChanges = Score.nlargest(5,'Change')[['Country', 'Change']]
topChanges
```

```python
downChanges = Score.nsmallest(5,'Change')[['Country', 'Change']]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.barh(topChanges['Country'],topChanges['Change'])
ax2.barh(downChanges['Country'],downChanges['Change'])
```
